```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import re

path = '/content/drive/MyDrive/IITP/sohyun/TimeSeriesAnomaly/codes/csv_files'
csvs = os.listdir(path)
csvs.sort()
```

```python
normal = [120, 136, 152, 166, 453, 550, 743, 898, 923]
abnormal = [248, 252, 260, 380, 394, 597, 598, 601, 604]

normal_csv = []
abnormal_csv = []

for csv in csvs:
  number = int(re.sub(r'[^0-9]', '', csv))

  if number in normal:
    normal_csv.append(pd.read_csv(path+"/"+csv))
  elif number in abnormal:
    abnormal_csv.append(pd.read_csv(path+"/"+csv))

flag = 1

for row in range(1, 10):
  plt.figure(figsize=(15,20))
  
  plt.subplot(9, 2, row*2-1)
  plt.title(normal[row-1])
  plt.plot(normal_csv[row-1])
  
  plt.subplot(9, 2, row*2)
  plt.title(abnormal[row-1])
  plt.plot(abnormal_csv[row-1])
# plt.plot(abnormal_csv[0])
```
